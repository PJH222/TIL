# 인구 과밀도 그려보기

```python
# 인구 과밀도 그려보기
city = ['서울','부산','울산','강릉','대전','대구']
lat = [30,32,33,31,34,38]
lon = [127,126,128,129,126.1,126.5]

pop_den = [11111,3000,2900,2700,1000,4400]
colors = ['r','g','b','c','m','k']

plt.scatter(lon,lat,s = np.array(pop_den),c=colors,alpha=0.5) # alpha 투명도를 의미
plt.xlabel("경도(longitude)")
plt.ylabel("위도(latitude)")
plt.title("집가고 싶다...")

for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
plt.show()
```

# 막대그래프
#### 필요한 부분만 체크하고 가자...

```python
import pandas as pd

data = {
    "01" : [27, 30],
    "02" : [15,22],
    "03" : [40,42],
    "04" : [33,37]
}
index_data = ["01","02","03","04"]
column_data = ["운동 시작 전", "운동 시작 후"]

df = pd.DataFrame(data,columns = index_data, index = column_data)
df

```

```python
IDs = ["01","02","03","04"]
before = [27,35,40,33]
after = [30,38,42,37]
```

```python
index = np.arange(len(IDs))
plt.bar(index, before)
plt.show()
# before = pd.Series(before)
# before.plot()
# 그렇게 예쁘게 그려지진 않는다...
```

```python
plt.bar(index,before , tick_label = IDs)
plt.show()
# 아까 보단 많이 나아졌다...
# 단순히 index가 아닌 IDs로 나타나서 좀더 통일성이 생김
```

```python
colors = ['r','g','y','b']
plt.bar(index,before,tick_label = IDs, color=colors)
plt.show()

# 이거 왤케 유치하게 그려지냐 ...
```

```python
plt.bar(index, before, color=colors, tick_label = IDs, width = 0.5)
```

```python
colors = ['r','g','b','m']
plt.barh(index,before, color=colors,tick_label=IDs) # horizontal 을 의미하는 h를 붙이면 옆으로 그래프가 눕는다 ...
plt.show()
```

```python
barWidth = 0.4
plt.bar(index,before,color='c',align = 'edge',width=barWidth, label='before') # 어떻게 보면 index = x , before = y 라고 이해하면 될 것 같다.
plt.bar(index + barWidth, after, color='b',align='edge', width = barWidth, label='after')
plt.xticks(index + barWidth, IDs)
plt.legend() # 왜하필 레전드얔ㅋㅋㅋ
plt.xlabel('회원 ID')
plt.ylabel('몸무게 변화')
plt.title('운동좀 하자~~~')
plt.show()
```

# 도수분포표 (히스토그램)

```python
import random
math_score = [random.randrange(5,101) for i in range(50)]
plt.hist(math_score)
# 별다른 설정값을 안넣어주면 계급은 10개로 알아서 나눠진다.
```

```python
plt.hist(math_score, bins = 8) # 이렇게 하면 계급이 8개로 바뀜
```

```python
plt.hist(math_score, bins=7)
plt.xlabel("수학 점수")
plt.ylabel("인원 수")
plt.title("수학 공부 좀 해라 이것들아~~")
plt.grid()

plt.show()
```
