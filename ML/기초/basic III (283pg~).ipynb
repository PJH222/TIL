# 산점도 그리기~~~

```python
height = [165,180,170,155,160,185]
weight = [40,80,75,45,55,60]

plt.scatter(height,weight)
plt.xlabel('Height(m)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid(True)
```

```python
import matplotlib.font_manager as fm

# 설치된 폰트 출력
font_list = [font.name for font in fm.fontManager.ttflist]
# print(font_list)
plt.rcParams['font.family'] = 'Gulim' # 대충 안에있는 굴림체로 쓰자...
```

```python
# 인구 과밀도 그려보기
city = ['서울','부산','울산','강릉','대전','대구']
lat = [30,32,33,31,34,38]
lon = [127,126,128,129,126.1,126.5]

pop_den = [11111,3000,2900,2700,1000,4400]
colors = ['r','g','b','c','m','k']

plt.scatter(lon,lat,s = np.array(pop_den),c=colors,alpha=0.5) # alpha 투명도를 의미
plt.xlabel("경도(longitude)")
plt.ylabel("위도(latitude)")
plt.title("집가고 싶다...")

for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
plt.show()
```
