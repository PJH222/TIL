```python
from pyspark import SparkConf , SparkContext
```

```python
conf = SparkConf().setMaster("local").setAppName("age_count")

sc = SparkContext(conf=conf)
sc
```

```python
filepath = "/home/ubuntu/working/spark/data/fakefriends.csv"
age = sc.textFile(f"file:///{filepath}")
age.collect()
```

```python
age_split = age.map(lambda x: x.split(","))
age_split.collect()
```

```python
gen = age_split.map(lambda x : (int(x[2])//10 * 10, x[0]))
```

```python
gen.collect()
```

```python
gen.groupByKey().collect()
```

```python
gen.groupByKey().mapValues(len).collect()
```
